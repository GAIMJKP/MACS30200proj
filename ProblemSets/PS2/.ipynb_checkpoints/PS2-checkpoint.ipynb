```python
prop_mat_inc = np.sum(bq_data, axis=0)
prop_mat_age = np.sum(bq_data, axis=1)
```

```python
SAMP = 100000
```

```python
age_probs = np.random.multinomial(SAMP, prop_mat_age)
income_probs = np.random.multinomial(SAMP, prop_mat_inc)
```

```python
age_probs
```

```python
age_freq = np.array([])
inc_freq = np.array([])
```

```python
for age, num_s in zip(age, age_probs):
    vec_age_s = np.ones(num_s)
    vec_age_s *= age
    age_freq = np.append(age_freq, vec_age_s)
```

```python
for pct, num_j in zip(pct, income_probs):
    vec_pct_j = np.ones(num_j)
    vec_pct_j *= pct
    inc_freq = np.append(inc_freq, vec_pct_j)
```

```python
data = np.vstack((age_freq, inc_freq))
```
